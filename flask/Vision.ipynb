```python
count = df_user_data['UNITID'].count()
```

```python
# selecting relevant columns for K-means clustering
cols = "ACTCMMID ACTENMID ACTMTMID ACT_AVG ADM_RATE_ALL SAT_AVG_ALL".split()
df_user_data_reduced = df_user_data[cols]
df_user_data_reduced.head()
```

```python
def k_value(row_count):
    if row_count > 0 and row_count < 5:
        return 1
    else:
        return 5
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = k_value(count))
```

```python
kmeans.fit(df_user_data_reduced)
```

```python
kmeans.cluster_centers_
```

```python
cluster_center = pd.DataFrame(kmeans.cluster_centers_, columns=cols)
```

```python
cluster_center
```

```python
# reducing college data dimensions to same as dimensions of user data
# preparing for least distance calculations
df_college_data = df_further_reduced[cols]
df_college_data.head()
```

```python
from sklearn.metrics import pairwise_distances_argmin_min
```

```python
recommendations = []
```

```python
for closest in range(5):
    """
    find closest college to cluster center
    add that college to recommendations list
    drop that college from dataframe so that same one does not get recommended again
    """
    closest, _ = pairwise_distances_argmin_min(kmeans.cluster_centers_, df_college_data)
    recommendations.append(df_college_data.iloc[closest])
    df_college_data = df_college_data.drop(df_college_data.index[closest])
```

```python
college_ids = []
for rec in recommendations:
    college_ids.append(rec.index)
```

```python
result = []
for college_id in college_ids:
    result.append(df_further_reduced.loc[college_id])
```

```python
res = []
for i in result:
    res.append(i.to_dict(orient='records')[0])
```
